Simulating Social Networks 

```python
import numpy as np
import pandas as pd

import networkx as nx

from pylab import rcParams
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
sb.set_style('whitegrid')
```

Generating a graph object

```python
dg = nx.gn_graph(7, seed=25)

for line in nx.generate_edgelist(dg, data=False):
    print(line)
```

Attaching Label to node

```python
dg.node[0]['name'] = 'Sam'
dg.node[0]['ABC'] = 'YES'
```

```python
print(dg.node[0])
```

```python
dg.node[0]['name'] = 'Frankie'
dg.node[1]['name'] = 'Mustafa'
dg.node[2]['name'] = 'Murtaza'
dg.node[3]['name'] = 'Saam'
dg.node[4]['name'] = 'Boy'
dg.node[5]['name'] = 'BooBoo'
dg.node[6]['name'] = 'Saafi'
```

```python
dg.add_nodes_from([
    (0,{'age':25}),
    (1,{'age':31}),
    (2,{'age':45}),
    (3,{'age':76}),
    (4,{'age':43}),
    (5,{'age':21}),
    (6,{'age':38})
])
```

```python
dg.add_nodes_from([
    (0,{'gender':'F'}),
    (1,{'gender':'M'}),
    (2,{'gender':'M'}),
    (3,{'gender':'M'}),
    (4,{'gender':'F'}),
    (5,{'gender':'M'}),
    (6,{'gender':'F'})
])
```

Drawing The Network

```python
nx.draw_circular(dg, node_color = 'blue', with_labels=True)
```

Visualization:

```python
labeldict = {0:'Frankie',1:'Mustafa',2:'Murtaza',3:'Saam',4:'Boy',5:'BooBoo',6:'Saafi'}
```

```python
nx.draw_circular(dg, labels = labeldict, node_color = 'yellow', with_labels=True)
```

```python
G = dg.to_undirected()
```

```python
nx.draw_spectral(G,labels=labeldict, node_color='yellow', with_labels=True)
```

```python
nx.draw_spectral(dg, labels = labeldict, node_color = 'yellow', with_labels=True)
```
